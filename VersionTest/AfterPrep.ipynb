```python
import tensorflow as tf
model = tf.keras.models.load_model('../VersionTest/multiclass_model')
```

```python
import numpy as np
from read_dataset import read_dataset
DATA_FILE = '../Datasets/Opportunities.csv'
dataset = read_dataset(DATA_FILE)
X_cols = dataset[['Amount','ExpectedRevenue','TotalOpportunityQuantity']]
y_cols = dataset[['AE','Alliances','BDR','Marketing']]

# Store the predictions on the test dataset
raw_predictions = model.predict(X_cols)

# post processing
final_predicted_ouput = []
for prediction in raw_predictions: # predictions is an array of array of strings
   final_predicted_ouput.append(
      {
      "AE":np.round_(prediction[0].astype(np.float64),3),
      "Alliances":np.round_(prediction[1].astype(np.float64),3),
      "BDR":np.round_(prediction[2].astype(np.float64),3),
      "Marketing":np.round_(prediction[3].astype(np.float64),3)
      })
   
# print your model output
print(final_predicted_ouput)
```
